```python

import time
import matplotlib.pyplot as plt

def iterative_method(n):
    sum_value = 0
    for i in range(1, n + 1):
        sum_value += i
    return sum_value

def recursive_method(n):
    if n == 0:
        return 0
    else:
        return n + recursive_method(n - 1)

N_values = [10, 100, 1000, 10000, 100000]

iterative_times = []
recursive_times = []

# Measure execution time for both methods
for N in N_values:
    start_time = time.time()
    iterative_method(N)
    iterative_times.append(time.time() - start_time)

    start_time = time.time()
    recursive_method(N)
    recursive_times.append(time.time() - start_time)

# Plot the results
plt.plot(N_values, iterative_times, label='Iterative')
plt.plot(N_values, recursive_times, label='Recursive')
plt.xscale('log')
plt.xlabel('N')
plt.ylabel('Time (seconds)')
plt.title('Time to find sum of first N natural numbers')
plt.legend()
plt.show()
```

```python
import time
import random
import matplotlib.pyplot as plt

# Function for linear search
def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1

# Function for binary search
def binary_search(arr, key):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == key:
            return mid
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1
    return -1

# Generate random array of 10000 elements
random_array = [random.randint(1, 1000) for _ in range(10000)]
random_array.sort()  # Sort the array for binary search

# Perform searches and plot the time taken
search_keys = [random.randint(1, 1000) for _ in range(5)]

linear_times = []
binary_times = []

for key in search_keys:
    start_time = time.time()
    linear_search(random_array, key)
    linear_times.append(time.time() - start_time)

    start_time = time.time()
    binary_search(random_array, key)
    binary_times.append(time.time() - start_time)

# Plotting the results
plt.plot(search_keys, linear_times, label='Linear')
plt.plot(search_keys, binary_times, label='Binary')
plt.xlabel('Search Key')
plt.ylabel('Time (seconds)')
plt.title('Time taken for linear and binary searches')
plt.legend()
plt.show()
```

```python
def string_to_integer(s):
    if len(s) == 0:
        return 0  # Base case: return 0 if the string is empty

    # Convert the first character of the string to an integer
    first_digit = int(s[0])

    # Recursively convert the remaining string to an integer
    integer_value = string_to_integer(s[1:])

    # Calculate the integer value
    result = first_digit * (10 ** (len(s) - 1)) + integer_value

    return result


input_string = "13531"  
result = string_to_integer(input_string)  # Convert the string to an integer
formatted_result = "{:,}".format(result)  # Format the result with thousand separators
print("Integer representation of '{}' is: {}".format(input_string, formatted_result))  
```

```python
def reverse_string(s):
    if len(s) <= 1:  
        return s
    else:
        # Recursively reverse the substring starting from the second character,
        return reverse_string(s[1:]) + s[0]

input_string = "pots&pans"  # The input string to be reversed
reversed_string = reverse_string(input_string) 
print("The reverse of '{}' is: {}".format(input_string, reversed_string))  
```

```python
def is_palindrome(s):
    
    if len(s) <= 1:
        return True  # Return True 
    # Check if the first and last characters are the same
    if s[0] != s[-1]:
        return False  
    
    return is_palindrome(s[1:-1])  # Check if the inner substring is a palindrome

input_string1 = "racecar"  
input_string2 = "gohangasalamiimalasagnahog" 

# Check if input strings are palindromes
print("Is '{}' a palindrome? {}".format(input_string1, is_palindrome(input_string1)))  
print("Is '{}' a palindrome? {}".format(input_string2, is_palindrome(input_string2)))  
```
